```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier

#import os
#for dirname, _, filenames in os.walk('/kaggle/input'):
#    for filename in filenames:
#        print(os.path.join(dirname, filename))

df_train = pd.read_csv('../input/titanic/train.csv')
df_test = pd.read_csv('../input/titanic/test.csv')
```

```python
df_train.update(df_train['Age'].fillna(df_train['Age'].mean()))
df_train = df_train.drop(['Cabin', 'Ticket', 'Name'], 1)
df_train.info()
```

```python
columns = df_train.columns
```

```python
df_train = df_train.dropna()
```

```python
target = df_train['Survived']
data = df_train.drop('Survived', 1)

data.info()
```

```python
dt_sex = dict(zip(data['Sex'].unique(),np.linspace(0,len(data['Sex'].unique())-1,len(data['Sex'].unique()))))
dt_emb = dict(zip(data['Embarked'].unique(),np.linspace(0,len(data['Embarked'].unique())-1,len(data['Embarked'].unique()))))
```

```python
dt_sex
```

```python
data = data.replace({'Sex': dt_sex})
data = data.replace({'Embarked': dt_emb})

data.head()
```

```python
decision_tree = DecisionTreeClassifier(random_state=0, max_depth=20)
decision_tree = decision_tree.fit(data, target)
```

```python
print(classification_report(target, decision_tree.predict(data)))
```

```python
df_test = df_test.drop(['Cabin', 'Ticket', 'Name'], 1)
```

```python
df_test = df_test.replace({'Sex': dt_sex})
df_test = df_test.replace({'Embarked': dt_emb})

df_test.info()
```

```python
df_test.update(df_test['Age'].fillna(df_test['Age'].mean()))
df_test.update(df_test['Fare'].fillna(df_test['Fare'].mean()))
```
